# 2020 Analysis

```python
import pandas as pd

filename = "../input/Michigan_2020_detailed_Votes.csv"
voting_df = pd.read_csv(filename)
```

```python
voting_df.head()
```

```python
working_df = voting_df.copy()
working_df['SPV_Trump_pct'] = working_df['SPV_Trump']*100/working_df['SPV_Total']
working_df['SPV_Biden_pct'] = working_df['SPV_Biden']*100/working_df['SPV_Total']

working_df['ICV_Trump_pct'] = working_df['ICV_Trump']*100/working_df['ICV_Total']
working_df['ICV_Biden_pct'] = working_df['ICV_Biden']*100/working_df['ICV_Total']

working_df['Trump_ICV_pct-SPV_pct'] = working_df['ICV_Trump_pct'] - working_df['SPV_Trump_pct']
working_df['Biden_ICV_pct-SPV_pct'] = working_df['ICV_Biden_pct'] - working_df['SPV_Biden_pct']

working_df.dropna(inplace=True)
```

```python
working_df.sort_values(by='Trump_ICV_pct-SPV_pct', ascending=False, inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

m, b = np.polyfit(working_df['SPV_Trump_pct'], working_df['Trump_ICV_pct-SPV_pct'],  1)

sns.scatterplot(data=working_df, x='SPV_Trump_pct', y='Trump_ICV_pct-SPV_pct');
plt.plot([0, 100], [m*0 + b, m*100+b], color='red', linestyle='dashed');

plt.yticks([-30, -20, -10, 0, 10, 20, 30]);
plt.xticks([0, 20, 40, 60, 80, 100]);
plt.title(f"TRUMP:   m={round(m, 2)},   b={round(b, 2)}");
```

```python
m, b = np.polyfit(working_df['SPV_Biden_pct'], working_df['Biden_ICV_pct-SPV_pct'],  1)

sns.scatterplot(data=working_df, x='SPV_Biden_pct', y='Biden_ICV_pct-SPV_pct');
plt.plot([0, 100], [m*0 + b, m*100+b], color='red', linestyle='dashed');

plt.yticks([-30, -20, -10, 0, 10, 20, 30]);
plt.xticks([0, 20, 40, 60, 80, 100]);
plt.title(f"BIDEN:   m={round(m, 2)},   b={round(b, 2)}");
```
